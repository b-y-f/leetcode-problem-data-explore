```python
lc_df = pd.read_json('out.json', lines=True)
lc_df = pd.json_normalize(lc_df['data'])

lc = lc_df.copy()
```

# Rank by like/total vote ratio

Get threshold total vote > 1000 and rank them...

```python
lc = lc[(lc['question.dislikes'] + lc['question.likes'])> 2000]
lc['like ratio'] = lc['question.likes']/(lc['question.dislikes'] + lc['question.likes'])
(lc
 .sort_values('like ratio', ascending=False)[[
     'question.questionId','question.title','question.isPaidOnly','question.difficulty','like ratio']]
 .head(20))
```
