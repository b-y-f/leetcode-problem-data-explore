```python
import pandas as pd
import numpy as np
```

```python
lc_df = pd.read_json('out.json', lines=True)
lc_df = pd.json_normalize(lc_df['data'])

lc = lc_df.copy()
```

```python
lc.head()
```

# Rank by like/total vote ratio

Get threshold total vote > 1000 and rank them...

```python
lc = lc[(lc['question.dislikes'] + lc['question.likes'])> 2000]
lc['like ratio'] = lc['question.likes']/(lc['question.dislikes'] + lc['question.likes'])
(lc
 .sort_values('like ratio', ascending=False)[[
     'question.questionId','question.title','question.isPaidOnly','question.difficulty','like ratio']]
 .head(20))
```

# Rank by difficulty

```python
def rank_by_diff(df):
    lc_easy = lc[lc['question.difficulty']=='Easy'].sort_values('like ratio', ascending=False)
    lc_med = lc[lc['question.difficulty']=='Medium'].sort_values('like ratio', ascending=False)
    lc_hard = lc[lc['question.difficulty']=='Hard'].sort_values('like ratio', ascending=False)
    return lc_easy,lc_med,lc_hard

easy,mid,hard = rank_by_diff(lc)
```
